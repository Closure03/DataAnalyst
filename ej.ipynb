```python
import bcrypt

def almacenar_contrasena():
  contrasena = input("ingrese su contraseña: ")
  sal = bcrypt.gensalt()
  hash = bcrypt.hashpw(contrasena.encode('utf-8'),sal)
  print("Si se almaceno")
  return hash, sal

def validar_contrasena(hash, sal):
  contrasena = input("Valide su contraseña: ")
  if bcrypt.checkpw(contrasena.encode('utf-8'), hash):
    print("contraseña correcta, Se dan accesos")
  else:
    print("contraseña incorrecta")


if __name__=="__main__":
  hash, sal = almacenar_contrasena()

  print("hash: ",hash)
  print("sal: ",sal)

  validar_contrasena(hash,sal)
```
